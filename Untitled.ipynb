```python
X=df2.drop("student_marks",axis="columns")
y=df2.drop("study_hours",axis="columns")
print("shape of X =",X.shape)
print("shape of y =",y.shape)
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df2,kind="reg")
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=51) 
print("shape of X_train =",X_train.shape)
print("shape of y_train =",y_train.shape)
print("shape of X_test =",X_test.shape)
print("shape of y_test =",y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, random_state=51)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
print("Predicted value:", y_pred)
```

```python
pd.DataFrame(np.c_[X_test,y_test,y_pred],columns=["study_hours","student_marks_original Percentage","student_marks_predicted Percentage"])
```

```python
rf.score(X_test,y_test)  
```

```python
plt.scatter(X_train,y_train)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X_test,y_test)
plt.plot(X_train,rf.predict(X_train),color="orange")
```
